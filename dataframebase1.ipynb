```python

import pandas as pd

# Ruta donde está tu archivo .txt (ajusta según tu caso)
ruta_archivo = 'base_1_transaccional.txt'


# Leer el archivo TSV y convertirlo en un DataFrame
df = pd.read_csv(ruta_archivo, sep='\t', encoding='utf-8')

# Renombrar las columnas para que coincidan con la estructura deseada
df_renombrado = df.rename(columns={
    'fecha_factura': 'fecha',
    'id_b2b': 'id',
    'municipio': 'municipio',
    'zona': 'zona',
    'categoria_b2b_macro': 'categoria_macro',
    'categoria_b2b': 'categoria',
    'subcategoria_b2b': 'subcategoria',
    'producto': 'producto',
    'valor_total': 'valor',
    'alineación con portafolio estratégico b2b': 'alineación con portafolio estratégico'
})

# Agregar columnas faltantes con valores nulos (puedes llenarlas después)
df_renombrado['pedido'] = None  # Placeholder para número de pedido
df_renombrado['edad'] = None    # Placeholder para edad del cliente
df_renombrado['asesor'] = None  # Placeholder para asesor
df_renombrado['punto_de_venta'] = None  # Placeholder para punto de venta
df_renombrado['cluster'] = None  # Placeholder para cluster
df_renombrado['color'] = None    # Placeholder para color
df_renombrado['cantidád'] = None  # Placeholder para cantidad
df_renombrado['precio'] = None   # Placeholder para precio (puedes calcularlo como valor / cantidád si tienes cantidad)

# Verifica el resultado
print("Primeras 5 filas del DataFrame reorganizado:")
print(df_renombrado.head())

# Muestra la información del DataFrame
print("\nInformación del DataFrame:")
print(df_renombrado.info())

# Opcional: Guarda el DataFrame en un nuevo archivo CSV
df_renombrado.to_csv('base_datos_reorganizada.csv', index=False)
print("DataFrame guardado como 'base_datos_reorganizada.csv'")
```
